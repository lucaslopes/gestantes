```python
import pandas as pd
from utils import get_path, group_cols
# from datatable import dt, f, as_type
```

# Descrição

Informações referentes a localizações em diversas escalas (Município, Regiões de Saúde, UF, Região).

# Fontes

## Municípios

```python
url_muns = 'https://raw.githubusercontent.com/kelvins/Municipios-Brasileiros/main/csv/municipios.csv'
path_muns = get_path('LOCALIDADES', 'municipios.csv.gzip')
# df_muns = pd.read_csv(url_muns)
# df_muns.to_csv(path_muns, index=False)
df_muns = pd.read_csv(path_muns)
df_muns
```

## Regiões de Saúde

```python
url_rsau = 'http://www.ans.gov.br/images/stories/noticias/planilha/regionais%20de%20sade.xls'
path_rsau = get_path('LOCALIDADES', 'regioes_saude.csv.gzip')
# df_rsau = pd.read_excel(url_rsau, header=2)
# df_rsau.to_csv(path_rsau, index=False)
df_rsau = pd.read_csv(path_rsau)
df_rsau
```

## Regiões do Brasil

```python
url_regs = 'https://www.estadosecapitaisdobrasil.com/'
path_regs = get_path('LOCALIDADES', 'regioes.csv.gzip')
# df_regs = pd.read_html(url_regs)[0]
# df_regs.to_csv(path_regs, index=False)
df_regs = pd.read_csv(path_regs)
df_regs
```

# Gerando tabela

## Merge

```python
df_loc = df_muns.copy()
df_loc['codigo_ibge'] = df_loc['codigo_ibge'].apply(lambda x: int(str(x)[:6]))
df_loc = df_loc.merge(df_rsau, how='left', left_on='codigo_ibge', right_on='CÓD. IBGE DO MUNICÍPIO')
df_loc = df_loc.merge(df_regs, how='left', left_on='UF', right_on='Sigla')
df_loc
```

## Ajustando Colunas

```python
columns = {
  'codigo_ibge': 'cod_municipio',
  'nome': 'nome_municipio',
  'capital': 'capital',
  'NOME DA REGIÃO DE SAÚDE': 'nome_regiao_saude',
  'codigo_uf': 'cod_uf',
  'UF': 'uf',
  'Estado': 'nome_uf',
  'Região': 'regiao',
  'latitude': 'latitude',
  'longitude': 'longitude',
}
```

```python
df_loc = df_loc[list(columns)]
df_loc = df_loc.rename(columns=columns)
df_loc
```

# Exportando tabela

```python
path_loc = get_path('LOCALIDADES', 'localidades.csv.gzip')
df_loc.to_csv(path_loc, index=False)
```
